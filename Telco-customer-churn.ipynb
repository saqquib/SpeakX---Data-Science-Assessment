```python
#indices of missing values
index_missing=np.array(df.isnull()).nonzero()[0] 
```

```python
y=df.Churn
X=df.drop(columns=['Churn','customerID'])
y=y.drop(index=index_missing)
```

```python
#label encoding 
X['gender']=X['gender'].map({'Male':1, 'Female':0})

binary_vars=['Partner','Dependents','PhoneService','PaperlessBilling']
X[binary_vars]=X[binary_vars].replace({'Yes':1, 'No':0})

X['tenure']=(X['tenure']-X['tenure'].mean())/X['tenure'].std()
X['MonthlyCharges']=(X['MonthlyCharges']-X['MonthlyCharges'].mean())/X['MonthlyCharges'].std()
X['TotalCharges']=(X['TotalCharges']-X['TotalCharges'].mean())/X['TotalCharges'].std()
```

```python
other_vars=['MultipleLines','InternetService','OnlineSecurity',
                 'OnlineBackup','DeviceProtection','TechSupport',
                 'StreamingTV','StreamingMovies','Contract','PaymentMethod']
onehot_encoded=pd.get_dummies(X[other_vars])
```
